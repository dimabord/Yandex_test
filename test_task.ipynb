#### Assumptions
1. Number of delivery zone vertexes is not defined, i.e. a delivery zone may be a triangle, a quadrangle, a pentagon et c.
2. The data itself is valid, i.e. at least 3 points are defined for each zone. Based on that assumption validation is not included

```python
import pandas as pd
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
from matplotlib import path
import numpy as np

#  pandas is used to read csv and convert it to numpy as it is faster than 
place_zone_coordinates = pd.read_csv('place_zone_coordinates.csv',
                                     names=['place_id', 'loc_lat', 'loc_lon', 'point_number']).to_numpy()

user_coordinates = pd.read_csv('user_coordinates.csv',
                               names=['user_id','loc_lat', 'loc_lon'])

points = user_coordinates[['loc_lat', 'loc_lon']].to_numpy()

results = pd.DataFrame(columns=['user_id','number_of_places_available'])
results.user_id = user_coordinates.user_id
results.number_of_places_available = 0
results = np.array(results.to_numpy(), dtype='int32')



for i in np.unique(place_zone_coordinates[:,0]):
    zone_coordinates = place_zone_coordinates[place_zone_coordinates[:,0]==i, 1:3]
    polygon = path.Path(zone_coordinates)
    results[:,1] += polygon.contains_points(points)
print(results)
pd.DataFrame(results, columns=['id', 'number_of_places_available']).to_csv('results.csv')
```
